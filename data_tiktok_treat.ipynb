```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
#Atribui a variável 'df' o csv 'tiktok_nlp_data'
tiktok = 'tiktok_nlp_data.csv'
tiktok_data = pd.read_csv(tiktok)
df = tiktok_data 
df
```

```python
#lista as colunas presentes no dataset
colunas = df.columns.tolist()
print(colunas)
```

```python
df.info()
#Verifica, nas colunas não-numéricas, quais os tipos de dados possíveis
print(df['claim_status'].unique())
print(df['verified_status'].unique())
print(df['author_ban_status'].unique())
```

```python
#Transforma os dados das colunas com duas opções em dados booleanos
mapear_claim = {'claim' : 1, 'opinion' : 0}
mapear_verif = {'verified' : 1, 'not verified' : 0}

df['claim_status'] = df['claim_status'].map(mapear_claim, na_action=None)
df['verified_status'] = df['verified_status'].map(mapear_verif, na_action=None)
```

```python
#print(df['claim_status'].unique())
print(df['verified_status'].unique())
```

```python
#Utiliza o método OneHotEncoder para a coluna com 3 opções de dado
encoder = OneHotEncoder(sparse=False)
one_hot_encoded = encoder.fit_transform(df[['author_ban_status']])
df_encoded = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(['author_ban_status']))
df = pd.concat([df, df_encoded], axis=1)
df.drop(columns='author_ban_status', inplace=True)
```

```python
#Normaliza os dados utilizando o método do Máximo Absoluto
from sklearn.preprocessing import MaxAbsScaler

scaler = MaxAbsScaler()
df_normalized = scaler.fit_transform(df)
colunas_normalizar = ['video_duration_sec', 'video_like_count', 'video_share_count', 'video_download_count', 'video_comment_count']
df[colunas_normalizar] = scaler.fit_transform(df[colunas_normalizar])
```
